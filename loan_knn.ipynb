```python
import numpy as numpy
import pandas as pd
df=pd.read_csv('/content/LoanApprovalPrediction.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.drop(['Loan_ID'], axis=1,inplace=True)
df
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Married']=le.fit_transform(df['Married'])
df['Education']=le.fit_transform(df['Education'])
df['Self_Employed']=le.fit_transform(df['Self_Employed'])
df['Property_Area']=le.fit_transform(df['Property_Area'])
df['Dependents']=le.fit_transform(df['Dependents'])
```

```python
df.isna().sum()
```

```python
x=df['LoanAmount'].mean()
print(x)
df['LoanAmount'].fillna(x,inplace=True)
print(df)
```

```python
x=df['Loan_Amount_Term'].mean()
print(x)
df['Loan_Amount_Term'].fillna(x,inplace=True)
print(df)
```

```python
x=df['Credit_History'].mean()
print(x)
df['Credit_History'].fillna(x,inplace=True)
print(df)
```

```python
df.info()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
x.ndim
```

```python
y.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
result=confusion_matrix(y_test,y_pred)
result
```

```python
from sklearn.metrics._plot.confusion_matrix import ConfusionMatrixDisplay
label=pd.unique(y_test)
cmd=ConfusionMatrixDisplay(result,display_labels=label)
cmd.plot()
```

```python
score=accuracy_score(y_test,y_pred)
score
```
